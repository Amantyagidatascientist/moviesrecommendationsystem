Fill the null values in 'spoken_languages' column with values from 'original_language' column

Fill null values in 'production_companies' with a specific value, e.g., 'Unknown'       
Fill null values in 'production_countries' with a specific value, e.g., 'Unknown'


```python
data['production_companies'].fillna('Unknown', inplace=True)
data['production_countries'].fillna('Unknown', inplace=True)
data['genres'].fillna('Unknown_genres', inplace=True)
data['overview'].fillna('Unknown', inplace=True)
```

To understand how your data is spread out or distributed

```python

columns_n=['vote_average','vote_count','revenue','runtime','budget','popularity']

plt.figure(figsize=(18, 24)) 

for i, col in enumerate(columns_n, 1):
    plt.subplot(6, 2, (i-1)*2 + 1)
    sns.kdeplot(data[col], fill=True)
    plt.title(f'KDE of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Density')
    plt.subplot(6, 2, (i-1)*2 + 2)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
